```python
data = p.read_csv('/content/drive/MyDrive/Colab Notebooks/Heart_Disease_Prediction.csv')
```

```python
data
```

```python
X= data[['Age','Sex','Chest pain type','BP','Cholesterol','FBS over 120','EKG results','Max HR','Exercise angina','ST depression','Slope of ST','Number of vessels fluro',
         'Thallium']]
Y= data['Heart Disease']
```

```python
data.head()
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.20)
```

```python
log_reg= LogisticRegression()
log_reg.fit(X_train,Y_train)
Y_pred=log_reg.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print ("Accuracy : ", accuracy_score(Y_test,Y_pred))
```
